# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
engine
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

from sqlalchemy import inspect

inspection = inspect(engine)

meas_columns = inspection.get_columns('measurement')
meas_columns

print("Measurement")
for column in meas_columns:
    print(column['name'], column["type"])
    
station_columns = inspection.get_columns('station')
station_columns

print("\nStations")
for column in station_columns:
    print(column['name'], column["type"])
```

```python
# Calculate the date 1 year ago from the last data point in the database

last_date = session.query(Measurement.date).\
            order_by(Measurement.date.desc()).first()
print(f"Last Date is: {last_date[0]}")

#Now we need to calculate the date 1 year go. Considering 2017 was a 365 day year, we'll subtract that many days

date_one_year_ago = (dt.datetime.strptime(last_date[0],'%Y-%m-%d') - dt.timedelta(days=365)).strftime('%Y-%m-%d')

print("-------------------------------")
print(f"Date from one year ago was: {date_one_year_ago}")
```

```python
# Perform a query to retrieve the data and precipitation scores

measurement_cols = (Measurement.date, Measurement.prcp)

precipitation_data = session.query(*measurement_cols).\
            filter(Measurement.date >= date_one_year_ago).all()

precipitation_data[0:6]
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column

precipitation_df = pd.DataFrame(precipitation_data, columns=['date','precipitation'])
precipitation_df.head()
```

```python
# Sort the dataframe by date

sorted_df = precipitation_df.sort_values(by = ['date'])
sorted_df.head()

sorted_df.set_index('date', inplace=True)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data

import matplotlib
import matplotlib.pyplot as plt

ax = sorted_df.plot(rot=90, figsize=(25,25))
ax.set_title("Precipitation Levels")

plt.savefig('Precipitation Levels.png')
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data

precipitation_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?

station_count = session.query(func.count(Station.id)).all()
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

count_by_station =  session.query(Measurement.station, func.count(Measurement.station)).\
                    group_by(Measurement.station).\
                    order_by(func.count(Measurement.station).desc()).all()
count_by_station
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

session.query(func.min(Measurement.tobs),func.max(Measurement.tobs),func.avg(Measurement.tobs)).\
        filter(Measurement.station == 'USC00519281').all()
```

```python
# Choose the station with the highest number of temperature observations.

highest_temp_data =  session.query(Measurement.date,Measurement.tobs).\
                    filter(Measurement.station == 'USC00519281').\
                    filter(Measurement.date > date_one_year_ago).all()

highest_temp_data[0:4]
```

```python
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

highest_temp_df= pd.DataFrame(highest_temp_data)
highest_temp_df.plot(kind="hist",bins=15,figsize=(20,10))

plt.xlabel("Temperature")
plt.title("data for station with highest no. of temperature observations")

plt.savefig("histogram for station.png")
```

## Bonus Challenge Assignment

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.


# Let's choose the date range 2/28/2016-3/5/2016
trip_start_date = '2016-02-28'
trip_end_date = '2016-03-05'

trip_data = (dt.datetime.strptime(last_date[0],'%Y-%m-%d') - dt.timedelta(days=365)).strftime('%Y-%m-%d')

def last_year_trip_data(start_date, end_date):
    
    hawaii_start_date = (dt.datetime.strptime(trip_start_date,'%Y-%m-%d') - dt.timedelta(days=365)).strftime('%Y-%m-%d')
    
    hawaii_end_date = (dt.datetime.strptime(trip_end_date,'%Y-%m-%d') - dt.timedelta(days=365)).strftime('%Y-%m-%d')
    
    return hawaii_start_date, hawaii_end_date

last_year_start_date, last_year_end_date = last_year_trip_data(trip_start_date, trip_end_date)

tmin, tavg, tmax = calc_temps(last_year_start_date, last_year_end_date)[0]
print(tmin, tavg, tmax)
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

plt.figure(figsize=(5,10))
plt.bar(x = "0", height=tavg, yerr=tmax-tmin,color='orange',alpha=.6,align='edge')
plt.ylim(0,100)
plt.xlabel("Avg Temp")
plt.ylabel("Temp (F)")
plt.title("Trip Avg Temp")

plt.savefig("Trip Avg Temp")
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.


amount_of_rainfall = [Station.station, Station.name, func.sum(Measurement.tobs), Station.latitude, Station.longitude, Station.elevation]

amount_of_rainfall_data = session.query(*amount_of_rainfall).\
                    group_by(Measurement.station).\
                    outerjoin(Station, Measurement.station==Station.station).\
                    filter(Measurement.date >= last_year_start_date).filter(Measurement.date <= last_year_end_date).\
                    order_by(func.sum(Measurement.tobs).desc()).all()

amount_of_rainfall_data

raninfall_df = pd.DataFrame(amount_of_rainfall_data,
                           columns=["Station","Name", "Total Rainfall", "Latitude", "Longitude", "Elevation"])

raninfall_df

# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation

#Considering that the maximum precipitation would be where maximum rainfall is:

raninfall_df.sort_values("Total Rainfall",ascending=False)
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# Set the start and end date of the trip

print(f"I had my Hawaii trip starting {last_year_start_date} and it ended on {last_year_end_date}")

# Use the start and end date to create a range of dates

hawaii_start = dt.datetime.strptime(last_year_start_date,'%Y-%m-%d')
hawaii_end = dt.datetime.strptime(last_year_end_date,'%Y-%m-%d') 
total_trip_time = hawaii_end - hawaii_start
normals = []

for i in range(total_trip_time.days + 1):
    day = hawaii_start + dt.timedelta(days=i)
    
    # Strip off the year and save a list of %m-%d strings
    every_day = dt.datetime.strftime(day,'%m-%d')
    normals.append(every_day)

# push each tuple of calculations into a list called `normals`

print("--------------------------------")
print(normals)
print("--------------------------------")

normal_data = []
for day in normals:
    normal_data.append(daily_normals(day)[0])
    
normal_data

# Loop through the list of %m-%d strings and calculate the normals for each date
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index

normals_df = pd.DataFrame(normal_data, columns=["tmin","tavg","tmax"], index=[normals])

normals_df.index = normals_df.index.get_level_values(0)

normals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
normals_df.plot(kind="area",stacked=False, alpha = 0.4,figsize=(10,10))
plt.xlabel("2016 travel date")
plt.ylabel("Temperature")

plt.savefig("Daily normals-Area plot")
```
